```python
#2 d scatter
iris.plot(kind='scatter',x='sepal_length',y ='sepal_width');
plt.grid()
plt.show()
```

```python
#2 d scatter plot with color coding for each flower type class
sns.set_style('whitegrid');
sns.FacetGrid(iris,hue='species',size=4)\
    .map(plt.scatter,'sepal_length','sepal_width')\
    .add_legend();
plt.show();
```

```python
plt.close();
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',size=3);
plt.show()
```

```python

#plot cdf of petal_length
iris = pd.read_csv('D:\datasets\IRIS.csv')
counts ,bins_edges=np.histogram(iris['petal_length'],bins = 10,density = True)
pdf= counts/sum(counts)
print(pdf)
print(bins_edges)
#compute cdf
cdf = np.cumsum(pdf)
plt.plot(bins_edges[1:],pdf)
plt.plot(bins_edges[1:],cdf)
plt.show()
```

```python
#mean ,variance, standard deviation
print('mean')
print(np.mean(iris['petal_length']))
print(np.mean(np.append(iris['petal_length'],50)))

print('\nstd-dev')
print(np.std(iris['petal_length']))
```

```python
#median 
print('\nmedian')
print(np.median(iris['petal_length']))
print(np.median(np.append(iris['petal_length'],50)))
```

```python
#QUANTILES
print('\nquantiles')
print(np.percentile(iris['petal_length'],(0,100,25)))
```

```python
#percentiles 
print('\n90th percentiles ')
print(np.percentile(iris['petal_length'],90))
```

```python
from statsmodels import robust
print('\nmedian abosulute deviation ')
print(robust.mad(iris['petal_length']))
```

```python
#box plot
sns.boxplot(x= 'species',y = 'petal_length',data =iris)
plt.show()  
```

```python
#violin plot
sns.violinplot(x= 'species',y = 'petal_length',data =iris,size = 8)
plt.show() 
```

```python
#multivariate 
#2 D scatter Density contors plot
sns.jointplot(x = 'petal_length', y = 'petal_width',data = iris)

plt.show()
```
